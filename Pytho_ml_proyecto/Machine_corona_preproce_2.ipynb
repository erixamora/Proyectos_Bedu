```python
import pandas as pd
```

```python
df = pd.read_csv('datos_actividad_biologica.csv')
df
```

```python
#limpiar df
df2 = df[df.standard_value.notna()]
df2
```

Preprocesamiento de los datos divir los compuestos en 3:
menores a 1000nM se considera activo
mayores a 10,000 nM se considera inactivo
y aquellos cuales sus valores esten entre 1,000 y 10,000 nM seran intermedio

```python
bioactivity_class= []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append("inactivo")
    elif float(i) <= 1000:
        bioactivity_class.append("activo")
    else:
        bioactivity_class.append("intermedio")
```

obtener el id de las moleculas que actuan en la proteina objetivo, molecule chembl, estos son las distintas moleculas, solo elegiremos 1
obtener molecule id, canonical smiles y el standad value

```python
selection = ['molecule_chembl_id' , 'canonical_smiles', 'standard_value']
df3 = df2[selection]
df3
```

```python
bioactivity_class =pd.Series(bioactivity_class, name= 'bioactivity_class')
df4= pd.concat([df3, pd.Series(bioactivity_class)], axis=1)

df4
```

```python
df4.to_csv('datos_biologicos_preprocesado.csv', index= False)
```
