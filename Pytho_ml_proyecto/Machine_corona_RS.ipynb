```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
df= pd.read_csv('Modelo_regresion_corona.csv')
df
```

```python
X = df.drop('pIC50', axis=1)
X
```

```python
Y = df.pIC50
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.feature_selection import VarianceThreshold
selection = VarianceThreshold(threshold=(.8 * (1 - .8)))    
X = selection.fit_transform(X)
```

```python
X.shape
```

80 20

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

Regression model

```python
import numpy as np
np.random.seed(100)
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
r2
```

```python
Y_pred = model.predict(X_test)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.set(color_codes= True)
sns.set_style("white")

ax = sns.regplot(Y_test, Y_pred, scatter_kws={'alpha':0.4}, color ='g')

ax.set_xlabel('pIC50 experimental', fontsize='large', fontweight='bold')
ax.set_ylabel('pIC50 predecido', fontsize='large', fontweight='bold')
ax.set_xlim(0, 12)
ax.set_ylim(0, 12)
ax.figure.set_size_inches(5, 5)
plt.show()

plt.savefig('plot_RL.pdf')
```
